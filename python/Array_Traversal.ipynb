2. Print Elements in Reverse Order -
    Input: [10, 20, 30, 40]
    Output: 40 30 20 10

```python
#Method-1
arr2 = [10, 20, 30, 40]
str2 = ""
for i in range(1,len(arr2)+1):
    str2 += str(arr2[-i])
    str2 += " "
print(str2)

#Method-2
print(" ".join(map(str,arr2[::-1])))
```

3. Print Even Numbers Only-
Input [3, 6, 9, 12, 15]
Output 6 12


```python
#Method-1
arr3 = [3, 6, 9, 12, 15]
str3 = ""
for i in arr3:
    if i%2 == 0:
        str3 += str(i)
        str3 += " "
print(str3)

#Method-2
x = lambda x : x%2 == 0
ar_list = " ".join(map(str,list(filter(x,arr3))))
# print(" ".join(map(str, ar_list)))
print(ar_list)
```

15. Find the Difference Between Sum of Even and Odd Elements :
Input [5, 6, 7, 8]
Output 2

```python
arr = [5,6,7,8]
even_sum = 0
odd_sum = 0
for i in range(len(arr)):
    if i%2==0:
        even_sum += arr[i]
    else:
        if i%2 !=0:
            odd_sum += arr[i]
m=max(even_sum,odd_sum)
min = min(even_sum,odd_sum)
print(m-min)
```

21. Find the Second Largest Element (Without Sorting) :
Input [10, 5, 8, 20, 15]

```python
arr = [10,5,8,20,15]
c=0
for i in arr:
    if i>c:
        c=i
li=[]
for j in arr:
    if j<c:
        li.append(j)
m=0
for k in li:
    if k>m:
        m=k
print(m)
```
