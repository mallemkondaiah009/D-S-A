```python
# Double Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if self.head is None:
            print("Empty Double Linked List!")
        else:
            temp = self.head
            while temp:
                print(temp.data, "-->", end=" ")
                temp = temp.next
            print("None")

    def insert_begining(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_ending(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    def insert_position(self, pos, data):
        new_node = Node(data)
        
        # Case 1: Insert at the beginning
        if pos == 1:
            self.insert_begining(data)
            return

        temp = self.head
        for _ in range(pos - 2):  # Traverse to the (pos-1)th node
            if temp is None:
                print(f"Position {pos} is out of bounds.")
                return
            temp = temp.next

        # Case 2: Insert at the end (if temp.next is None after traversal)
        if temp is None or temp.next is None:
            self.insert_ending(data)
        else:
            # Case 3: Insert in the middle
            new_node.next = temp.next
            new_node.prev = temp
            temp.next.prev = new_node
            temp.next = new_node


# Create the doubly linked list and populate it
L1 = DoublyLinkedList()
n1 = Node(100)
L1.head = n1
n2 = Node(200)
n2.prev = n1
n1.next = n2
n3 = Node(300)
n3.prev = n2
n2.next = n3
n4 = Node(400)
n4.prev = n3
n3.next = n4

# Insert nodes
L1.insert_begining(50)
L1.insert_begining(10)
L1.insert_ending(500)
L1.insert_ending(600)
L1.insert_position(2, 25)  # Insert 25 at position 2
L1.insert_position(1, 5)   # Insert 5 at position 1

L1.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if self.head is None:
            print("Empty Double Linked List!")
        else:
            temp = self.head
            while temp:
                print(temp.data, "-->", end=" ")
                temp = temp.next
            print("None")

    def insert_begining(self, data):
        new_node = Node(data)
        if self.head is None:  # Handle empty list
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_ending(self, data):
        new_node = Node(data)
        if self.head is None:  # Handle empty list
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    def insert_position(self, pos, data):
        if pos < 1:
            print("Invalid position!")
            return

        if pos == 1:  # Insert at the beginning
            self.insert_begining(data)
            return

        new_node = Node(data)
        temp = self.head

        # Traverse to the node just before the position
        for _ in range(pos - 2):
            if temp is None:  # Position out of range
                print("Position out of range!")
                return
            temp = temp.next

        if temp.next is None:  # Insert at the end
            self.insert_ending(data)
        else:  # Insert in the middle
            new_node.next = temp.next
            new_node.prev = temp
            temp.next.prev = new_node
            temp.next = new_node


# Example usage
L1 = DoublyLinkedList()
n1 = Node(100)
L1.head = n1
n2 = Node(200)
n2.prev = n1
n1.next = n2
n3 = Node(300)
n3.prev = n2
n2.next = n3
n4 = Node(400)
n4.prev = n3
n3.next = n4
L1.insert_begining(50)
L1.insert_begining(10)
L1.display()
```
