```python
li=[]
for i in range(1,5+1):
    for j in range(1,5+1):
        li.append(j)
for k in range(1,5+1):
    print(*li[:k], '==>', sum(li[:k]))
```

```python
n=5
for i in range(n):
    for j in range(n):
        print((i+j) % n+1,end=' ')
    print()
```
