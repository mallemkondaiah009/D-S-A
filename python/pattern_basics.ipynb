üß± LEVEL 1 ‚Äî Row & Column Awareness (foundation)

First rule of all patterns:

i = row number
j = column number


```python
n=4
for i in range(n):
    for j in range(n):
        print(i+1,end=" ")
    print()
```

```python
n=4
for i in range(n):
    for j in range(n):
        print(j+1,end=" ")
    print()
```

üìê LEVEL 2 ‚Äî Diagonal (slanted logic)

```python
n=4
for i in range(n):
    for j in range(n):
        print(i+j+1, end=" ")
    print()
```

```python
n=4
for i in range(n):
    for j in range(n):
        print(i-j+1, end=" ")
    print()
```

```python
n=4
for i in range(n):
    for j in range(n):
        print(n-i-j, end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(i+j+1, end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(n-i+j, end=" ")
    print()
```

üîÅ LEVEL 3 ‚Äî Wrapping (cyclic patterns)

```python
n=5
for i in range(n):
    for j in range(n):
        print((i+j)%n+1, end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print((i-j)%n+1, end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print((n-1-i+j)%n+1, end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print((n-i-j), end=" ")
    print()
```
